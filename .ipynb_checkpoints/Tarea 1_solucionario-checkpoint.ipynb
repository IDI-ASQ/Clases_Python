**Ejercicio 1**

Escribir un programa que lea un entero porsitivo, $n$, introducido por el usuario y después muestre en pantalla la suma de todos los enteros al cuadrado desde 1 hasta $n$. La suma de los $n$ primeros enteros cuadrados positivos puede ser calculada de la siguiente forma: 

$$ \mbox{suma_cuadrado} = \frac{n(n+1)(2n+1)}{6} $$

*Introduzca un número entero: 5* <br>
*La suma de los primeros números enteros cuadrados desde 1 hasta 5 es 55.0*

```python
n  = int(input())
suma_cuadrado = n*(n+1)*(2*n+1)/6
print('El resultado es {}'.format(suma_cuadrado))
```

1. Importe el la data ``diabetes.csv``
2. Cree una nueva columna llamada ``nueva_columna``. Dicha columna debe contener la raiz cuadrada del campo ``age``
2. Luego elimine la columna ``Pregnancies``
3. Agrupe ``group by`` por Outcome, el promedio de las columnas ``Glucose``, `` Age`` . ``nueva_columna``

```python
import pandas as pd
```

```python
df = pd.read_csv('data/diabetes.csv')
```

```python
df
```

```python
df['nueva_columna'] = df['Age'].apply(lambda x: x**(1/2))
df
```

```python
df.drop(labels = ['Pregnancies'], axis = 1, inplace =True)
```

```python
df
```

```python
df.groupby(['Outcome'])[['Glucose','Age','nueva_columna']].mean()
```
